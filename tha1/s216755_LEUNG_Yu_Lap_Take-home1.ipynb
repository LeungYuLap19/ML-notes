```python
# TODO: Add import statements
import pandas as pd
from sklearn.linear_model import LinearRegression

# Assign the Pandas dataframe to this variable.
# TODO: Load the data
bmi_life_data = pd.read_csv('bmi_and_life.csv')

# Make and fit the linear regression model
#TODO: Fit the model and Assign it to bmi_life_model
x = bmi_life_data[['BMI']]
y = bmi_life_data['Life expectancy']
bmi_life_model = LinearRegression().fit(x, y)

# Make a prediction using the model
# TODO: Predict life expectancy for a BMI value of 23.47563 and print the result
laos_life_exp = bmi_life_model.predict([[23.47563]])
print(laos_life_exp)
```

```python
# TODO: Add import statements
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Assign the data to predictor and outcome variables
# TODO: Load the data.
# TODO: Make sure that you've split out the data into the predictor feature in X and outcome feature in y
# TODO: For X, make sure it is in a 2-d array of 20 rows by 1 column. You might need to use NumPy's reshape function to accomplish this.
train_data = pd.read_csv('Polynomial_Regression_Data.csv')
train_data.sort_values(by='Var_X', inplace=True)
X = train_data[['Var_X']].values
y = train_data['Var_Y'].values

X = X.reshape(-1, 1)

# Create polynomial features
# TODO: Create a PolynomialFeatures object 'poly_feat', then fit and transform the predictor feature. Pay attention to how to set the degree of features.
poly_feat = PolynomialFeatures(degree=8)
# TODO: Create the polynomial features 'X_poly' by using the PolynomialFeatures object's .fit_transform() method
X_poly = poly_feat.fit_transform(X)

# Make and fit the polynomial regression model
# TODO: Create a LinearRegression object and fit it to the polynomial predictor features
poly_model = LinearRegression()
poly_model.fit(X_poly, y)

# The following code help you visualize the model
plt.scatter(X, y, color = "red") 
plt.plot(X, poly_model.predict(X_poly), color = "blue") 
plt.show()

# Hints: Sort the data in the csv file by 'Var_X' in ascending order
# Otherwise, multiple lines will be displayed
```
